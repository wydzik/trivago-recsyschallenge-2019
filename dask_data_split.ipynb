```python
import pandas as pd
import numpy as np
import dask.dataframe as dd
```

```python
dask_data_set = dd.read_csv("C:/Users/Wydzik/Downloads/trivagoRecSysChallengeData2019_v2/train.csv")
```

```python
dask_no_duplicate = dask_data_set.copy().compute()
```

```python
dask_no_duplicate
```

```python
dask_no_duplicate = dask_no_duplicate.drop_duplicates(subset = ["session_id","step"]).compute()
```

```python
dask_no_duplicate
```

```python
len(dask_no_duplicate.session_id.unique())
```

```python
dask_unique_session_ids = dask_no_duplicate.session_id.unique()
dask_unique_sessions = len(dask_unique_session_ids)
dask_unique_sessions
```

```python
divider = 0.8
```

```python
dask_number_of_train_sessions = int(dask_unique_sessions * divider)
dask_number_of_train_sessions
```

```python
dask_last_train_session_id = dask_unique_session_ids[dask_number_of_train_sessions]
dask_last_train_session_id
```

```python
dask_last_training_index = dask_no_duplicate[dask_no_duplicate.session_id == dask_last_train_session_id][-1:]
dask_last_training_index
```

```python
# wyszło mi z ręcznych obliczeń, że 0.8 setu jest w 27 partycji,
# więc teraz trzeba sprawdzić, czy wyjdzie ta sama wartość, co wyżej, bo nie do końca wiem jak to działa w dasku
dask_train_partition = int(dask_no_duplicate.npartitions * divider)
dask_train_partition
```

```python
dask_end_of_training = dask_no_duplicate.get_partition(dask_train_partition).compute()
```

```python
dask_end_of_training[dask_end_of_training.session_id == dask_last_train_session_id]
```

```python
last_training_index = dask_end_of_training[dask_end_of_training.session_id == dask_last_train_session_id][-1:].index[0]
last_training_index
# Wyszło to samo :D czyli git
```

```python
end_of_training = dask_end_of_training.loc[dask_end_of_training.index <= last_training_index]
```

```python
end_of_training
```

```python
dask_beginning_of_test = dask_end_of_training.loc[dask_end_of_training.index > last_training_index]
dask_beginning_of_test
```

```python
dask_new_train_set = dask_no_duplicate.partitions[:dask_train_partition].append(end_of_training).compute()
dask_new_train_set
```

```python
#zanim połącze testowe ze sobą, muszę sprawdzić jak to wygląda z tym podziałem.
dask_no_duplicate[dask_no_duplicate.session_id == "266f553b63fe8"].compute()
```

```python
#czyli najlepiej by było to połączyć przed resztą pliku (chyba)
temp = dask_no_duplicate.partitions[dask_train_partition + 1 :]
dask_new_test_set = dd.concat([dask_beginning_of_test,temp]).compute()
#z tym, że z tego mi się teraz zrobił dataframe nie podzielony, więc pewnie się powtarzają indexy. 
#Ale może to nie szkodzi, zobaczyymy
#(a może nie trzeba było tego tak łączyć, tylko wystarczyło appendem dodać partycję do pozostałych)
```

```python
dask_new_test_set
```

```python
dask_blurred_test = dask_new_test_set.copy()
dask_blurred_test = dask_blurred_test.reset_index(drop = True)
dask_blurred_test
```

```python
last_step_indexes =  dask_blurred_test[dask_blurred_test.step == 1].index - 1
last_index = dask_blurred_test.tail(1).index[0]
last_step_indexes = np.append(last_step_indexes, last_index)
#no i teraz widzę, że może być problem, bo tam w środku pewnie się powtarzają.
#ale jak się zresetowało indexy to od razu jest lepiej.
#teraz pokazuje -1 ale nie szkodzi, po potem -1 i tak nie spełni warunku, to niech sobie będzie.
#za to wiem, że trzeba dodać ostatni indeks, ponieważ on jest siłą rzeczy ostatnim, ale po nim nie ma pierwszego
#więc nie spełniał powyższego warunku
last_step_indexes
```

```python
records_to_blur = dask_blurred_test[(dask_blurred_test.index.isin(last_step_indexes)) & (dask_blurred_test.action_type == "clickout item")]
records_to_blur
```

```python
indexes_to_blur = dask_blurred_test[(dask_blurred_test.index.isin(last_step_indexes)) & (dask_blurred_test.action_type == "clickout item")].index
indexes_to_blur
```

```python
dask_blurred_test.loc[(dask_blurred_test.index.isin(indexes_to_blur)),'reference'] = np.nan
```
